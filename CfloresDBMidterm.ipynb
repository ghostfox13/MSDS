```python
dfs = dfs.drop(['INCIDENT_NUMBER','OFFENSE_CODE','OFFENSE_DESCRIPTION','DISTRICT','REPORTING_AREA','UCR_PART','Location'], axis=1)
```

```python
print(dfs.dtypes)
dfs['OCCURRED_ON_DATE'] = pd.to_datetime(dfs['OCCURRED_ON_DATE'])
dfs['MONTH'] = dfs['OCCURRED_ON_DATE'].dt.month_name()
```

```python
#Creating time frames breaking up the 24 hour day in 4 time blocks
dfs['TIME_FRAME'] = np.where((dfs['OCCURRED_ON_DATE'].dt.hour >= 0) & (dfs['OCCURRED_ON_DATE'].dt.hour < 6),
                                 '00:00 - 6:00am',
                        np.where((dfs['OCCURRED_ON_DATE'].dt.hour >= 6) & (dfs['OCCURRED_ON_DATE'].dt.hour < 12),
                                 '6:01am - 12:00',
                        np.where((dfs['OCCURRED_ON_DATE'].dt.hour >= 12) & (dfs['OCCURRED_ON_DATE'].dt.hour < 18),
                                 '12:00 - 18:00',
                        np.where((dfs['OCCURRED_ON_DATE'].dt.hour >= 18) & (dfs['OCCURRED_ON_DATE'].dt.hour < 24),
                                 '18:00 - 00:00', 'unspecified'))))
```

Once data is cleaned, breakdown of crime rate trends will be presented, starting with year, month, day, and finally timeblocks of 6 hour windows. 

```python
fig, (ax1, ax2, ax3, ax4) = plt.subplots(4, 1, figsize=(10,10))

# Yearly crime trends
year_ct = []
for i in dfs.YEAR.unique():
    year_ct.append(len(dfs[dfs['YEAR']==i]))
sns.pointplot(x=dfs.YEAR.unique(),y=year_ct,color='red', ax=ax1)
ax1.set_xlabel('Year')
ax1.set_ylabel('Crime Count')
ax1.set_title('Yearly Crime Trends 2015-2022')

# Monthly crime trends
month_ct = []
for i in dfs.MONTH.unique():
    month_ct.append(len(dfs[dfs['MONTH']==i]))
sns.pointplot(x=dfs.MONTH.unique(),y=month_ct,color='blue', ax=ax2)
ax2.set_xlabel('Month')
ax2.set_ylabel('Crime Count')
ax2.set_title('Monthly Crime Trends 2015-2022')

# Daily crime trends
day_ct = []
for i in dfs.DAY.unique():
    day_ct.append(len(dfs[dfs['DAY']==i]))
sns.pointplot(x=dfs.DAY.unique(),y=day_ct,color='green', ax=ax3)
ax3.set_xlabel('Day')
ax3.set_ylabel('Crime Count', fontsize=12)
ax3.set_title('Daily Crime Trends 2015-2022')

#hourly crime trends
hour_ct = []
for i in dfs.TIME_FRAME.unique():
    hour_ct.append(len(dfs[dfs['TIME_FRAME']==i]))
sns.pointplot(x=dfs.TIME_FRAME.unique(),y=hour_ct,color='yellow', ax=ax4)
ax4.set_xlabel('Hours')
ax4.set_ylabel('Crime Count')
ax4.set_title('Hourly Crime Trends 2015-2022')

plt.tight_layout()
plt.show()
sns.set(font_scale=0)
```
