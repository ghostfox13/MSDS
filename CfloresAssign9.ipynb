```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from xgboost import XGBRegressor
from sklearn.metrics import mean_squared_error
```

```python
df = pd.read_csv('ToyotaCorolla.csv')
df = df.drop(columns=['Id', 'Model', 'Fuel_Type', 'Met_Color'])
train_df, val_df = train_test_split(df, test_size=0.3, random_state=42)

X_train = train_df.drop(columns=['Price'])
y_train = train_df['Price']
X_val = val_df.drop(columns=['Price'])
y_val = val_df['Price']
```

```python
lr_model = LinearRegression()
lr_model.fit(X_train, y_train)

xgb_model = XGBRegressor(n_estimators=1000, learning_rate=0.05, random_state=42)
xgb_model.fit(X_train, y_train)

lr_pred = lr_model.predict(X_val)
lr_rmse = np.sqrt(mean_squared_error(y_val, lr_pred))

xgb_pred = xgb_model.predict(X_val)
xgb_rmse = np.sqrt(mean_squared_error(y_val, xgb_pred))

print("Linear Regression RMSE:", lr_rmse)
print("XGBoost RMSE:", xgb_rmse)
```

```python
from sklearn.inspection import plot_partial_dependence
import matplotlib.pyplot as plt
```

```python
feature_idx = X_train.columns.get_loc('Mfg_Year')

fig, ax = plt.subplots(figsize=(8, 5))
plot_partial_dependence(lr_model, X_train, features=[feature_idx], ax=ax, line_kw={'linewidth': 2})
plt.title("Linear Regression: Partial Dependence Plot for Mfg_Year")
plt.xlabel("Mfg_Year")
plt.ylabel("Price")
plt.tight_layout()
plt.show()

fig, ax = plt.subplots(figsize=(8, 5))
plot_partial_dependence(xgb_model, X_train, features=[feature_idx], ax=ax, line_kw={'linewidth': 2})
plt.title("XGBoost: Partial Dependence Plot for Mfg_Year")
plt.xlabel("Mfg_Year")
plt.ylabel("Price")
plt.tight_layout()
plt.show()
```

```python
ice_samples = X_val.sample(10, random_state=42)
```

```python
feature = 'Mfg_Year'
```

```python
feature_range = np.linspace(X_train[feature].min(), X_train[feature].max(), 100)
```

```python
preds = np.zeros((10, len(feature_range)))
```

```python
for i, (idx, sample) in enumerate(ice_samples.iterrows()):
    for j, value in enumerate(feature_range):
        sample_copy = sample.copy()
        sample_copy[feature] = value
        
        preds[i, j] = lr_model.predict(sample_copy.values.reshape(1, -1))
```

```python
plt.plot(feature_range, preds[i], alpha=0.5, label=f'Sample {i+1}')
plt.xlabel(feature)
plt.ylabel('Price')
plt.title(f'ICE plot for {feature} (Linear Regression)')
plt.legend()
plt.show()
```
