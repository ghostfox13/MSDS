```python
data = df.drop("death",axis=1)
target = df["death"]
```

```python
from sklearn.model_selection import train_test_split
# Split the data into training and test dfs using 80/20 split
data_train, data_test, target_train, target_test = train_test_split(data, target, test_size=0.2, random_state=42)
```

## LogisticRegression

```python
from sklearn.linear_model import LogisticRegression
logisticReg = LogisticRegression(random_state=42).fit(data_train, target_train)
logisticReg_predict = logisticReg.predict(data_test)
```

```python
from sklearn.metrics import accuracy_score,recall_score,f1_score,precision_score,roc_auc_score
df_metrix = pd.DataFrame(columns=['Model','SMOTE','HyperParameter','accuracy_score','roc_auc_score','f1_score','precision_score'])
def get_metrics(df,target_test,predict,model_name,smote_flag='No',hyper_parameter_flag='No'):
    accuracyScore = accuracy_score(target_test, predict)
    roc_aucScore= roc_auc_score(target_test, predict)
    f1Score=f1_score(target_test, predict)
    precisionScore = precision_score(target_test, predict)
   
    print(f"accuracy_score {accuracyScore}")
    print(f"roc_auc_score {roc_aucScore}")
    print(f"f1_score {f1Score}")
    print(f"precision_score {precisionScore}")

    d2 = pd.DataFrame([{'Model':model_name,'SMOTE':smote_flag,'HyperParameter':hyper_parameter_flag,
                        'accuracy_score':accuracyScore,'roc_auc_score':roc_aucScore,'f1_score':f1Score,
                        'precision_score':precisionScore}])
    df = pd.concat([df,d2],ignore_index=True)
    return df
```

```python
df_metrix = get_metrics(df_metrix,target_test,logisticReg_predict,'LogisticRegression')
```

```python
df_metrix
```

```python
from sklearn.model_selection import GridSearchCV
# Define hyperparameters to tune
hyperparameters = {"penalty": ["l1", "l2", "elasticnet"],
                   "C": [0.001, 0.01, 0.1, 1, 10, 100]}

# Define logistic regression model
lr = LogisticRegression(random_state=42)

# Apply GridSearchCV to tune hyperparameters
lr_tuned = GridSearchCV(lr, hyperparameters)
lr_tuned.fit(data_train, target_train)

# Get best hyperparameters and apply on test set
best_lr = lr_tuned.best_estimator_


lr_tuned.best_estimator_
```

```python
lr_pred = best_lr.predict(data_test)
df_metrix = get_metrics(df_metrix,target_test,lr_pred,'LogisticRegression','No','Yes')
```
